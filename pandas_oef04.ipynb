# Oplossingen
## Oefening 1
Het is een Series, dus je kunt maar 1 waarde meegeven met .at. Je mag ook niet vergeten om vierkante haakjes te gebruiken. Het juiste antwoord is dus ser['Kathleen'] (antwoord 5)
## Oefening 2
.at gebruik je zoals een NumPy array: [rij, kolom], maar dan met indexen en kolomnamen. Dus df_k3.at['Kristel', 'geboortedatum'] (antwoord 2)
## Oefening 3
De indexkolom telt niet mee, dus de geboortedatum staat in kolom 1: .iat[1, 1] (antwoord 4)
## Oefening 4
De index is een RangeIndex, maar bij .loc is het eindpunt altijd inbegrepen, dus print(df_k3.loc[0:1, 'voornaam':'achternaam']) (antwoord 3)
## Oefening 5
Bij .iloc is het eindpunt niet inbegrepen. Er waren dus twee mogelijkheden: print(df_k3.iloc[1:3, 0:2]) en print(df_k3.iloc[1:, 0:2])
## Oefening 6 
Alhoewel de volgende code ook werkt:
```
df_k3.loc['Josje':'Klaasje', 'land'] = 'Nederland'
```
Is het toch veiliger om advanced indexing te gebruiken. Je wil immers de waarden wijzigen voor Josje en Klaasje. Afhankelijk van de volgorde van de rijen is dat niet hetzelfde als 'van Josje tot Klaasje':
```
df_k3.loc[['Josje','Klaasje'], 'land'] = 'Nederland'
```
## Oefening 7
Dit is in mijn ogen de beste oplossing:
```
df_k3.loc[['Kathleen', 'Josje', 'Klaasje'], 'land'] = ['Belgie', 'Nederland', 'Nederland']
```
alhoewel je natuurlijk ook .iloc kunt gebruiken:
```
df_k3.iloc[-3:, -1] = ['België', 'Nederland', 'Nederland']
```
## Oefening 8
Wanneer je bij de vorige oefeningen al advanced indexing had gebruikt, was dit niet zo moeilijk:
```
df_k3.loc[['Karen', 'Klaasje'], 'land'] = ['België', 'Nederland']
```
## Oefening 9
Aangezien we maar 1 waarde moeten wijzigen, kunnen we .at/.iat gebruiken. Aangezien we index kennen ('Klaasje') en de kolomnaam ('land) is .at waarschijnlijk de gemakkelijkste oplossing:
```
df_k3.at['Klaasje', 'land'] = 'Nederland'
```
## Oefening 10
Je hebt hier twee mogelijkheden. Aangezien we geen indexnaam meer hebben ('Klaasje' kunnen we niet gebruiken), is .iat waarschijnlijk het gemakkelijkst. We moeten immers de laatste kolom van de laatste record wijzigen:
```
df_k3.iat[-1, -1] = 'Nederland'
```
Maar .at zou ook werken. Hou er wel rekening mee dat je hier voor de index geen -1 kunt gebruiken:
```
df_k3.at[4, 'land'] = 'Nederland'
```
