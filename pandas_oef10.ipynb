In het volgende voorbeeld hebben beide dataframes dezelfde kolomnaam. Wat gebeurt er wanneer je de kolommen van beide dataframes samenvoegt met pd.concat()?

1. Je krijgt een foutmelding omdat kolomnamen uniek moeten zijn
1. De kolom van de tweede dataframe krijgt de naam 'telefoon_1' 

```python
import pandas as pd
voornamen = ['Karen', 'Kristel', 'Kathleen']
vaste_telefoon = ['03/534.23.12', '015/32.12.56', '27 71 142 26 73']
gsm = ['0478/12.30.22', '0477/86.29.12', pd.NA]
df1 = pd.DataFrame({'telefoon':vaste_telefoon},dtype=pd.StringDtype(), index=voornamen)
df2 = pd.DataFrame({'telefoon': gsm}, dtype=pd.StringDtype(), index=voornamen)
df1.info()
df2.info()
```
