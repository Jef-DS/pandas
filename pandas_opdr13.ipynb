# Opdracht multiindex
We beginnen met de 'most streamed spotify songs' van 2024 te downloaden.

```python
# Streaming data van Spotify
from pathlib import Path
from zipfile import ZipFile
import requests
URL='https://www.kaggle.com/api/v1/datasets/download/nelgiriyewithana/most-streamed-spotify-songs-2024'
SPOTIFY_CSV = 'Most Streamed Spotify Songs 2024.csv'
SPOTIFY_ZIP = 'spotify_2024.zip'
spotify_path = Path(SPOTIFY_CSV)
if not spotify_path.exists():
    data = requests.get(URL)
    with open(SPOTIFY_ZIP, 'wb') as f:
        f.write(data.content)
    with ZipFile(SPOTIFY_ZIP) as fzip:
        fzip.extractall()
```
