# Opdracht merging
De data voor deze opdracht is afkomstig van de [UC Irvine Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Auto+MPG)

## Download de data

```python
from pathlib import Path

import requests
URL1= 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv'
URL2= 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv'
CARS1_CSV = 'cars1.csv'
CARS2_CSV = 'cars2.csv'
if not Path(CARS1_CSV).exists():
    data = requests.get(URL1)
    with open(CARS1_CSV, 'wb') as f:
        f.write(data.content)
if not Path(CARS2_CSV).exists():
    data = requests.get(URL2)
    with open(CARS2_CSV, 'wb') as f:
        f.write(data.content)
```

## Genereer 400 random users

```python
random = requests.get('https://randomuser.me/api/?results=400')
random.json()
```

## Maak een dataframe
Maak een dataframe met de kolommen 'voornaam' en 'achternaam' op basis van random.json(). De namen zullen verschillen omdat jij andere namen gedownload hebt. Maar je moet een dataframe maken met kolommen voornaam en achternaam. De index is een gewone RangeIndex.
We kunnen pd.json_normalize gebruiken om de hiërarchische json-data te 'flatten'

```python
import pandas as pd
users_json = random.json()
users_df = pd.json_normalize(users_json['results']).loc[:, ['name.first', 'name.last']]
users_df.columns = ['voornaam', 'achternaam']
users_df
```

## We kunnen het ook zelf doen op de Python manier


```python
import pandas as pd
users_json = random.json()
users = []
for item in users_json['results']:
    voornaam = item['name']['first']
    achternaam = item['name']['last']
    users.append({'voornaam': voornaam, 'achternaam': achternaam})
users_df = pd.DataFrame(users)
users_df
```

## Importeer de twee car datasets en bekijk ze
Lees de twee datasets en bewaar ze in de variabelen *cars1* en *cars2*

```python
import pandas as pd
CARS1_CSV = 'cars1.csv'
CARS2_CSV = 'cars2.csv'
cars1 = pd.read_csv(CARS1_CSV)
display(cars1.head())
cars2 = pd.read_csv(CARS2_CSV)
display(cars2.head())
```

## Probleem met cars1
Het bestand cars1 bevat een aantal rare kolommen. Zorg dat je unnamed kolommen verwijdert

```python
cars1 = cars1.loc[:, :'car']
cars1.head()
```

## Hoeveel records zijn er in elke dataset?

```python
print('records in cars1:',cars1.shape[0])
print('records in cars2:', cars2.shape[0])
```

## Zet de gegevens van cars1 en cars2 in één dataset
Beide datasets bevatten dezelfde kolommen. Je kunt de rijen dus samenvoegen. Zet beide datasets samen in een nieuwe dataframe *cars*

```python
cars = pd.concat([cars1, cars2], axis='index')
print('Aantal records in cars: ', cars.shape[0])
cars.head()
```

## Maak kolom 'owners' in cars
Vertrek van onderstaande *random generator* om *owner-id's* te maken tussen 0 en 400. Zorg ervoor dat je evenveel id's genereert als er rijen zijn in het dataframe *cars*. Voeg die *owner-id's* toe als extra kolom aan het dataframe *cars*.

```python
import numpy as np
rng = np.random.default_rng(42)
user_ids = rng.integers(0, 400, cars.shape[0])
cars['owners'] = user_ids
cars.tail(20)
```

## Voeg cars en users_df samen
De kolom *owners* in cars is een (in databank termen) een *foreign key* naar de RangeIndex in users_df. Voeg beide datasets samen om basis van die 'foreign key relatie'. De naam van het nieuwe dataframe is *carusers_df*

```python
carusers_df = pd.merge(cars, users_df, left_on='owners', right_index=True).drop(columns='owners')
```

Wanneer je *carusers_df* toont, zou je de onderstaande structuur moeten krijgen. De waarden in *voornaam* en *achternaam zullen natuurlijk verschillen omdat jij andere namen gedownload hebt.
