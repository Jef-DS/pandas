## Begin met de download

```python
from pathlib import Path
from zipfile import ZipFile
import requests
URL= 'https://www.kaggle.com/api/v1/datasets/download/openfoodfacts/world-food-facts'
FOOD_FACT_ZIP = 'food_facts.zip'
food_path = Path(FOOD_FACT_ZIP)
if not food_path.exists():
    data = requests.get(URL)
    with open(FOOD_FACT_ZIP, 'wb') as f:
        f.write(data.content)
    with open(FOOD_FACT_ZIP, 'rb') as f:
        ZipFile(f).extractall()
```
