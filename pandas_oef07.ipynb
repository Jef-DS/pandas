# Wanneer werd er meer ingezet?
Het dataframe bevat de inzetten van 1978 tot december 2025. Er zijn dus gegevens van de 20ste eeuw en van de 21ste eeuw.
1. Maak twee dataframes: df_20ste met de gegevens van vóór 1 januari 2000 en df_21ste met de gegevens vanaf 1 januari 2000
1. Bereken de gemiddelde inzet voor beide dataframes
1. Wanneer er gemiddeld meer ingezet (op basis van deze getallen): in de 20ste of in de 21ste eeuw?

# Oplossing

```
import openpyxl
STATISTIEKEN = 'statistieken-lotto-12-25.xlsx'
werkboek = openpyxl.load_workbook(STATISTIEKEN)
sheets = werkboek.sheetnames
print(f'{sheets=}')
import pandas as pd
df = pd.read_excel(STATISTIEKEN, sheet_name=sheets[4], skiprows=2, nrows=10)
df.info()
df
begin_21ste_eeuw = pd.to_datetime('2000-01-01')
df_20ste = df[df['Trekkingsdatum'] < begin_21ste_eeuw]
df_21ste = df[df['Trekkingsdatum'] >= begin_21ste_eeuw]
gemiddelde_20ste = df_20ste['TOP 10 inzetten'].mean()
gemiddelde_21ste = df_21ste['TOP 10 inzetten'].mean()
if gemiddelde_20ste > gemiddelde_21ste:
    print('Er werd meer ingezet in de 20ste eeuw')
else:
    print('Er werd meer ingezet in de 21ste eeuw')
```
