## Download van de data
We kunnen de data downloaden van [deze URL](https://www.kaggle.com/api/v1/datasets/download/markmedhat/euro-2012-stats-team)

```python
from pathlib import Path
from zipfile import ZipFile
import requests
URL='https://www.kaggle.com/api/v1/datasets/download/markmedhat/euro-2012-stats-team'
EURO_STATS_ZIP = 'euro-2012-stats.zip'
euro_path = Path(EURO_STATS_ZIP)
if not euro_path.exists():
    data = requests.get(URL)
    with open(EURO_STATS_ZIP, 'wb') as f:
        f.write(data.content)
    with open(EURO_STATS_ZIP, 'rb') as f:
        ZipFile(f).extractall()
```
