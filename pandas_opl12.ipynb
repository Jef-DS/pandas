# Opdracht timeseries
Download de volgende data. Zet ze in een dataframe df_funds. Zorg ervoor dat 'Date' de index wordt en herkend wordt als een datum. Toon een overzicht van de kolommen

```python
import pandas as pd
URL = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'
df_funds = pd.read_csv(URL, parse_dates=['Date'], index_col='Date')
df_funds.info()
```

Toon de eerste 20 records van df_funds

```python
df_funds.head(20)
```

De gegevens zijn niet gesorteerd op de datum. Sorteer op datum en bewaar het resultaat in df_funds. Toon de eerste 20 records en kijk of ze inderdaad correct gesorteerd zijn.

```python
df_funds = df_funds.sort_index()
df_funds.head(20)
```

# Sommeer de Municipal Bond
Sommeer de Municipal Bond waarden telkens  op het einde van de maand.

```python
df_municipal = df_funds['Municipal Bond'].resample('ME').sum()
df_municipal
```

Zitten er negatieve waarden in df_municipal? Wat is trouwens de gemiddelde waarde?

```python
# zorg voor een functie waarmee je een idee kunt krijgen over de waarden in df_municipal
df_municipal.describe()
```

Zie je een verschil tussen *to_period* en *resample*?

```python
df_funds['Period'] = df_funds['Municipal Bond'].index.to_period('M')
df_funds.groupby('Period')['Municipal Bond'].sum()
```

```python
df_funds['Period'] = df_funds['Municipal Bond'].index.to_period('M')
df_funds.groupby('Period')['Municipal Bond'].sum()
```

Plot de waarden van df_municipal

```python
df_municipal.plot()
```

Bereken het 7-maands gemiddelde voor df_municipal. Zet beide waarden (df_municipal en het 7-maands gemiddelde in een plot)

```python
import matplotlib.pyplot as plt
df_municipal_7_maands = df_municipal.rolling(7).mean()
plt.plot(df_municipal.index.year, df_municipal, label='Municipal data')
plt.plot(df_municipal_7_maands.index.year, df_municipal_7_maands, label='7-daags gemiddelde')
plt.legend()
plt.show()
```
