# Opdracht groeperen en pivoteren
Om te laten zien dat je data niet eerst moet downloaden: je kunt ook een URL gebruiken met read_csv

```python
import pandas as pd
URL = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
df = pd.read_csv(URL, delimiter='|')
df.head()
```

## Zorg voor een index
Lees de gegevens opnieuw, maar nu met user_id als indexkolom. Bewaar het resultaat in een variabele 'users'

```python
users = pd.read_csv(URL, delimiter='|', index_col='user_id')
users
```

## Welke kolommen zijn er en wat zijn de types
Toon kolommen en hun types. Kijk meteen of er NA-waarden zijn

```python
users.info()
```

## Toon statistieken
Toon voor alle kolommen (dus ook voor de niet-numerieke) de statistieken

```python
users.describe(include='all')
```

## Voor welke kolommen zou je een categorietype kunn instellen
Stel een categorietype in voor twee kolommen (ik hoop dat je ziet welke)

```python
users['gender'] = users['gender'].astype('category')
users.occupation = users.occupation.astype('category')
users.info()
```

## Gemiddelde leeftijd per occupation
Wat is de gemiddelde leeftijd voor elke occupation (zorg ervoor dat je geen FutureWarning krijgt)

```python
users.groupby('occupation', observed=False).mean('age')
```

## Onehot encoding
Maak bijkomende kolommen 'man' en 'vrouw'. De kolom 'man' moet een 1 bevatten voor gender='M' en een 0 voor gender='F'. En de kolom 'vrouw' moet een 0 bevatten voor gender='M' en een 1 voor gender='F'

```python

users['man'] = 0
users.loc[users.gender == 'M', 'man'] = 1
users['vrouw'] = 0
users.loc[users.gender == 'M', 'vrouw'] = 1
users
```

## Percentage mannen voor elk beroep
Toon het percentage mannen voor elk beroep, sorteer van groot naar klein

```python
(users.groupby('occupation', observed=False).man.sum() / users.groupby('occupation', observed=False).man.count()).sort_values(ascending=False) * 100
```

## oudste en jongste
Toon de hoogste en de laagste leeftijd per occupation

```python
users.groupby('occupation', observed=False).age.agg(['min', 'max'])
```
